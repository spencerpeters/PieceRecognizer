```python
def plotHistogram(image, channel):
    channelImage = image[:, :, channel]
    histo, _ = np.histogram(channelImage, bins=np.arange(256))
    plt.plot(histo)
```

```python
plotHistogram(hsv, 1)
```

```python
plotHistogram(hsv, 2)
```

```python
plotHistogram(bn, 0); plotHistogram(bn, 1); plotHistogram(bn, 2)
```
