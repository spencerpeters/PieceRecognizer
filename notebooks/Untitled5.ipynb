```python
def downsampleCleanAndPlot(img):
    down = img[0:-1:downsamplingFactor, 0:-1:downsamplingFactor, :]
    denoised = denoise_nl_means(down)
    fig, axes = plt.subplots(2, 2, figsize=(15, 15))
    axes = axes.flatten()
    axes[0].imshow(down)
    axes[1].imshow(denoised)
    origHisto, _ = np.histogram(rgb2gray(down), range=(0, 1), bins=256)
    denoisedHisto, _ = np.histogram(rgb2gray(denoised), range=(0,1), bins=256)
    axes[2].plot(origHisto)
    axes[3].plot(denoisedHisto)
```

```python
downsampleCleanAndPlot(sgr2)
```

```python
downsampleCleanAndPlot(skimage.img_as_float(bn))
```
